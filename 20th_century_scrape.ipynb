```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# This automatically manages the driver version for you
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
```

```python
# Define the URL
url = "https://en.wikipedia.org/wiki/Key_events_of_the_20th_century"

# Use Selenium to load the page
driver.get(url)

# Grab the page source to parse with BeautifulSoup
from bs4 import BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Extract all text
scraped_text = soup.get_text()
```

```python
with open('20th_century_key_events.txt', 'w', encoding='utf-8') as f:
    f.write(scraped_text)
```
