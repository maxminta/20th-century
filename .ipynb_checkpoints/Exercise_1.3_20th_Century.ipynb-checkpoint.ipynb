```python
import pandas as pd

# Creating the data with the name 'Books'
data = pd.DataFrame({
    "Author": ["Tolkien", "Orwell", "Woolf", "Hemingway"],
    "Books": [15, 9, 12, 11],
    "Genre": ["Fantasy", "Dystopian", "Modernist", "Realist"]
})

data
```

```python
# Multiplying the 'Books' column by 10
data['Points'] = data['Books'].apply(lambda x: x * 10)
data
```

```python
# Multiplying Books by 2 using lambda
data['Double_Books'] = data['Books'].apply(lambda x: x * 2)
data
```

```python
# If Genre is 'Fantasy', label it 'Magic', else 'Reality'
data['Setting'] = data['Genre'].apply(lambda x: 'Magic' if x == 'Fantasy' else 'Reality')
data
```
